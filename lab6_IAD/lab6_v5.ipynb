```python
#PI-211, Turevskyi D.V., Variant: 5 

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Параметри
n = 1000  
mean = [0, 1]  
R = np.array([[1, -0.9], [-0.9, 2]])  

# Генерація даних
data = np.random.multivariate_normal(mean, R, n)
x, y = data[:, 0], data[:, 1] 
print(data)
```

```python
# Візуалізація на площині
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")
sns.scatterplot(x=x, y=y, color='b', alpha=0.6, s=60, edgecolor="w", linewidth=0.5)
#sns.regplot(x=x, y=y, scatter=False, color='r', ci=None)
plt.title('Графік розсіювання з лінією регресії', fontsize=16)
plt.xlabel('Ознака x', fontsize=14)
plt.ylabel('Ознака y', fontsize=14)
plt.grid(True)
plt.show()
```

```python
# 2. Обчислення коефіцієнта кореляції Пірсона
pearson_corr,_ = stats.pearsonr(x, y)

print(f"Коефіцієнт кореляції Пірсона: {pearson_corr:.4f}")

```

```python
# 3. Перевірка гіпотези про некорельованість ознак
_, p_value = stats.pearsonr(x, y)
a = 0.05  
if p_value < a:
    print(f"Гіпотеза відхиляється, p-значення: {p_value:.4f}")
else:
    print(f"Гіпотеза приймається, p-значення: {p_value:.4f}")

# 4. Порівняння з "реальним" коефіцієнтом кореляції
# Реальний коефіцієнт кореляції для моделі з матриці коварацій R
real_corr = R[0, 1] / (np.sqrt(R[0, 0]) * np.sqrt(R[1, 1]))
print(f"Реальний коефіцієнт кореляції Пірсона: {real_corr:.4f}")
```
