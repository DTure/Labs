```python
#PI-211, Turevskyi D.V., Variant: 5 

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Параметри
n = 90  
c = 4   

# Генерація даних
x, y = make_blobs(n_samples=n, centers=c, n_features=2, random_state=38)
```

```python
# 1.Візуалізація даних з класами
plt.figure(figsize=(12, 6))
sns.scatterplot(x=x[:, 0], y=x[:, 1], hue=y, palette='Set2', s=100, edgecolor='k')
plt.title('Кластери даних')
plt.xlabel('Ознака 1')
plt.ylabel('Ознака 2')
plt.grid(True)
plt.show()
```

```python
# 2. Алгоритм k-середніх (кластеризація)
kmeans = KMeans(n_clusters=c, random_state=38)
clusters = kmeans.fit_predict(x)
```

```python
# 3. Візуалізація результатів кластеризації(центри кластерів)
plt.figure(figsize=(12, 6))
sns.scatterplot(x=x[:, 0], y=x[:, 1], hue=clusters, palette='Set2', s=100, edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='w', marker='X', edgecolor="black", label='Центри кластерів')
plt.title('Кластери з центрами')
plt.xlabel('Ознака 1')
plt.ylabel('Ознака 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 4. Дендрограма 
linked = linkage(x, 'ward')
plt.figure(figsize=(12, 6))
dendrogram(linked, 
           orientation='top', 
           labels=y, 
           distance_sort='descending', 
           show_leaf_counts=True,
           leaf_rotation=0,  
           #color_threshold=0
           )
plt.title('Дендрограма')
plt.xlabel('Індекс')
plt.ylabel('Відстань')
plt.grid(True)
plt.show()
```
